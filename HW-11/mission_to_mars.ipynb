```python
# Print data from marsweath_url to make sure there is data 

print(marsweath_resp.text)
```

```python
# Mars Facts URL to Scrape

marsfacts_url = 'https://space-facts.com/mars/'
```

```python
# Retrieve data from nasa and store in nasa_resp 

marsfacts_resp = requests.get(marsfacts_url)
```

```python
# Print data from nasa_url to make sure there is data 

print(marsfacts_resp.text)
```

```python
# Create BeautifulSoup object

soup = bs(marsfacts_resp.text, 'html.parser')
```

```python
# Find the div that will allow you to retrieve the news title and news paragraph
# Use this html, do not go back to the website to look for it

marsfacts_results = soup.find_all('table', class_='tablepress tablepress-id-mars')
```

```python
# print only the parts of the html that have div, class_=slide since that is where the news title and news paragraph is

print(marsfacts_results)
```

```python

# Loop through results to retrieve mars facts
# Print the results to verify it is correct

for result in marsfacts_results:
    #mars_facts = result.find('table', class_='tablepress tablepress-id-mars').text
    ediameter = result.find('tr', class_='row-1 odd').text
    pdiameter = result.find('tr', class_='row-2 even').text
    mass = result.find('tr', class_='row-3 odd').text
    moons = result.find('tr', class_='row-4 even').text
    orbitdist = result.find('tr', class_='row-5 odd').text
    orbitper = result.find('tr', class_='row-6 even').text
    surfacetemp = result.find('tr', class_='row-7 odd').text
    firstrec = result.find('tr', class_='row-8 even').text
    recordby = result.find('tr', class_='row-9 odd').text
    print(ediameter)
    print(pdiameter)
    print(mass)
    print(moons)
    print(orbitdist)
    print(orbitper)
    print(surfacetemp)
    print(firstrec)
    print(recordby)
                                    

    # Dictionary to be inserted into MongoDB
    post = {
        'Equatorial Diameter': ediameter,
        'Polar Diameter': pdiameter,
        'Mass': mass,
        'Moons': moons,
        'Orbit Distance': orbitdist,
        'Orbit Period': orbitper,
        'Surface Temperature': surfacetemp,
        'First Record': firstrec,
        'Recorded By': recordby,
    }
```
