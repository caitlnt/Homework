```python
# import libraries and dependencies

import time
from splinter import Browser
from bs4 import BeautifulSoup as bs
import requests
import pandas as pd
import numpy as np
from selenium import webdriver
```

```python
# Nasa Mars News URL to Scrape

nasa_url = 'https://mars.nasa.gov/news/'
```

```python
# Retrieve data from nasa

html = requests.get(nasa_url)

# Create BeautifulSoup object; parse with 'html.parser'
soup = bs(html.text, 'html.parser')
```

```python
# Print data from nasa_url to make sure there is data 

print(soup.prettify())
```

```python
# Find the div that will allow you to retrieve the news title
# Use this html, do not go back to the website to look for it
# Latest news title

news_title = soup.find_all("div", class_="content_title")[0].text

print(news_title)
```

```python
# Find the div that will allow you to retrieve the news paragraph
# Use this html, do not go back to the website to look for it
# Latest news paragraph

news_paragraph = soup.find_all("div", class_="rollover_description_inner")[0].text

print(news_paragraph)
```

```python
# Mars Hemispheres URL to Scrape

# Using Splinter to find current Featured Mars Image on JPL url
# JPL Mars Space Images url

executable_path = {'executable_path': 'C:\Chrome\chromedriver'}
browser = Browser('chrome', **executable_path, headless=False)
mhemi_url = 'https://astrogeology.usgs.gov/search/results?q=hemisphere+enhanced&k1=target&v1=Mars'
browser.visit(mhemi_url)
```

```python
# Retrieve latest full mars images

html = browser.html
soup = bs(html, 'html.parser')
```

```python
# Find the div that will allow you to retrieve the latest mars images

mhemi_image = soup.find_all('div',class_="item")
```

```python
# get list of URLs for each hemisphere
mhemi_img_url_text = []

for item in mhemi_image:
        # Use Beautiful Soup's find() method to navigate and retrieve attributes
        
        link = item.find('a')
        href = link['href']
             
        mh_url = ('https://astrogeology.usgs.gov' + href)

        mhemi_img_url_text.append(mh_url)
```

```python
# run for loop going through each url and getting the title and sample url
# put values in as dictionary
hemisphere_image_urls = []

for url in mhemi_img_url_text:
    
        browser.visit(mh_url)
        html = browser.html
        soup = bs(html, 'html.parser')
        
        titles = soup.find('h2',class_="title")
    
        browser.click_link_by_text('Sample')
        
        img = browser.windows[0].next.url
        
        urls = {
            'title':titles.text,
            'mhemi_img_url':img
        }
         
        hemisphere_image_urls.append(urls)
        
        print('-----------')
        print(titles.text)
        print(img)
```

```python
# Mars Facts URL to Scrape

marsfacts_url = 'https://space-facts.com/mars/'
```
